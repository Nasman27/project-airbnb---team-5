```python
df3= pd.get_dummies(df_dropped3, columns=['room type', 'cancellation_policy', 'instant_bookable'])
#trying to figure out how to change categorical data into numerical data for EDA
```

```python
df3.head()
```

the above is  one hot encoding method. It takes the values in each col [room type, cancellation policy, and instant bookable] and turns it into a column.
The new column has a one if in the original dataframe, there was a value of that specifc category in the column.
to break it down further: room type had values entire home/apt, hotel room, private room, and shared room. With one hot encoding it takes the room column, splits it into four columns of zeros and ones. The four new columns are named after the
values in room type. The new columns have a one if that value was in the room type column originally. 
Purpose: so we can see if there is any correlation between the different room types/cancellation policy/insta bookable and bookings

```python
#construction year by neighbourhood group (NG) summary stats

df_construction_year_ng_mean = df3.groupby('neighbourhood group')['Construction year'].mean()
df_construction_year_ng_median = df3.groupby('neighbourhood group')['Construction year'].median()
df_construction_year_ng_var = df3.groupby('neighbourhood group')['Construction year'].var()
df_construction_year_ng_std = df3.groupby('neighbourhood group')['Construction year'].std()
df_construction_year_ng_sem = df3.groupby('neighbourhood group')['Construction year'].sem()
df_construction_year_ng_max = df3.groupby('neighbourhood group')['Construction year'].max()
df_construction_year_ng_min = df3.groupby('neighbourhood group')['Construction year'].min()

df_construction_year_summary_ng = pd.DataFrame({"Construction Year Mean": df_construction_year_ng_mean,
                                            "Construction Year Median": df_construction_year_ng_median,
                                            "Construction Year Variance": df_construction_year_ng_var,
                                            "Construction Year Std Dev": df_construction_year_ng_std,
                                            "Construction Year SEM": df_construction_year_ng_sem,
                                            "Construction Year Max": df_construction_year_ng_max,
                                            "Construction Year Min": df_construction_year_ng_min})

df_construction_year_summary_ng
```

```python
#construction year by neighbourhood (N) summary stats

df_construction_year_n_mean = df3.groupby('neighbourhood')['Construction year'].mean()
df_construction_year_n_median = df3.groupby('neighbourhood')['Construction year'].median()
df_construction_year_n_var = df3.groupby('neighbourhood')['Construction year'].var()
df_construction_year_n_std = df3.groupby('neighbourhood')['Construction year'].std()
df_construction_year_n_sem = df3.groupby('neighbourhood')['Construction year'].sem()
df_construction_year_n_max = df3.groupby('neighbourhood')['Construction year'].max()
df_construction_year_n_min = df3.groupby('neighbourhood')['Construction year'].min()


df_construction_year_summary_n = pd.DataFrame({"Construction Year Mean": df_construction_year_n_mean,
                                            "Construction Year Median": df_construction_year_n_median,
                                            "Construction Year Variance": df_construction_year_n_var,
                                            "Construction Year Std Dev": df_construction_year_n_std,
                                            "Construction Year SEM": df_construction_year_n_sem,
                                            "Construction Year Min": df_construction_year_n_min,
                                            "Construction Year Max": df_construction_year_n_max})

df_construction_year_summary_n
```
