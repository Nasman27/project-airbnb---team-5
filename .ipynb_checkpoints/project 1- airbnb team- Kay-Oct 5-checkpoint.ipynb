```python
#Average Price Per Neighbourhood Group

data.groupby(['neighbourhood group'])['price'].mean().plot(kind='bar', color='darkturquoise')
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")
plt.xlim(-1,5)
plt.ylim(0,710)
plt.title("Average Price Per Neighbourhood Group")

plt.show()
```

```python
#average price by neighbourhood group (NG) summary stats

average_price_mean = data.groupby('neighbourhood group')['price'].mean()
average_price_median = data.groupby('neighbourhood group')['Construction year'].median()
average_price_var = data.groupby('neighbourhood group')['Construction year'].var()
average_price_std = data.groupby('neighbourhood group')['Construction year'].std()
average_price_sem = data.groupby('neighbourhood group')['Construction year'].sem()
average_price_max = data.groupby('neighbourhood group')['Construction year'].max()
average_price_min = data.groupby('neighbourhood group')['Construction year'].min()

average_price_summary = pd.DataFrame({"Price Mean": average_price_mean,
                                            "Price Median": average_price_median,
                                            "Price Variance": average_price_var,
                                            "Price Std Dev": average_price_std,
                                            "Price SEM": average_price_sem,
                                            "Price Max": average_price_max,
                                            "Price Min": average_price_min})

average_price_summary
```

```python
df3= pd.get_dummies(df_dropped3, columns=['room type', 'cancellation_policy', 'instant_bookable'])
#trying to figure out how to change categorical data into numerical data for EDA
```

```python
df3.head()
```

the above is  one hot encoding method. It takes the values in each col [room type, cancellation policy, and instant bookable] and turns it into a column.
The new column has a one if in the original dataframe, there was a value of that specifc category in the column.
to break it down further: room type had values entire home/apt, hotel room, private room, and shared room. With one hot encoding it takes the room column, splits it into four columns of zeros and ones. The four new columns are named after the
values in room type. The new columns have a one if that value was in the room type column originally. 
Purpose: so we can see if there is any correlation between the different room types/cancellation policy/insta bookable and bookings
